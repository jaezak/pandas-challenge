```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete = complete.head(21)
complete
```

```python
#average math score
average_math = student_data["math_score"].mean()
average_math
```

```python
# average reading score
average_reading = student_data["reading_score"].mean()
average_reading
```

```python
# % passing math (the percentage of students who passed math)
pass_math = student_data["math_score"]
pass_math = student_data["math_score"]>=70
pass_math
```

```python
# * % passing reading (the percentage of students who passed reading)
pass_reading = student_data["reading_score"]
pass_reading = student_data["reading_score"]>=70
pass_reading
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_pass = student_data[(pass_math) & (pass_reading)]
overall_pass
```
